```python
#data loading
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("ai_job_dataset.csv")
df.head(10)
```

```python
#data cleaning deleted rows containg Null values
df.dropna(inplace=True)
print(df)
```

```python
df = df.dropna()  
print(df)
```

```python
# data transformations companies with more job posts 
top_companies = df['company_name'].value_counts().head(10)
print(top_companies)
```

```python
#creating constant data types
import matplotlib.pyplot as plt
df['salary'] = pd.to_numeric(df['salary_usd'], errors='Null')
df['posting_date'] = pd.to_datetime(df['posting_date'], errors='Null') 
```

```python
#sorting and displaying top 5 rated scores
top_rated = df.sort_values('benefits_score', ascending=False).head(5)
print(top_rated)
```

```python
# Histogram of salary
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
plt.hist(df['salary'], bins=30, color='skyblue', edgecolor='red')
plt.title('Distribution of AI Salaries (USD)')
plt.xlabel('Salary USD')
plt.ylabel('Frequency')
plt.show()
```

```python
#pie chart
top_countries = df['employee_residence'].value_counts().head(5)
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'orange']
plt.figure(figsize=(6,6))
plt.pie(top_countries, labels=top_countries.index, autopct='%1.1f%%',
        colors=colors, explode= [0.1, 0, 0, 0, 0]  , shadow=True)
plt.title('Top 5 Countries by AI Job Count')
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(8,5))
plt.scatter(df['years_experience'], df['salary_usd'], alpha=0.6, c='purple', edgecolors='w', s=50)
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary USD')
plt.show()
```
